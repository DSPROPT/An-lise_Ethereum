# PARTE 1: TEORIA
Vamos comear com uma breve introduo a estes conceitos.

O desvio padro  uma medida da quantidade de variao ou disperso de um conjunto de valores. Um baixo desvio padro indica que os valores tendem a estar prximos da mdia do conjunto, enquanto um alto desvio padro indica que os valores esto dispersos numa gama mais ampla.

Por outro lado, a distribuio normal, tambm conhecida como curva de sino,  uma funo que representa a distribuio de muitos conjuntos de dados na natureza e na sociedade, incluindo finanas.

## PARTE 2: CDIGO PYTHON
Agora, vamos ver como podemos usar Python para obter os dados do Ethereum, calcular o desvio padro e gerar uma distribuio normal. Para isso, vamos utilizar a biblioteca yfinance, que nos permite extrair dados financeiros do Yahoo Finance, e as bibliotecas numpy e scipy para os nossos clculos e anlises estatsticas.

##Primeiro, precisamos de obter os dados do Ethereum.




```python
# Importamos as bibliotecas necessrias


try:
    import yfinance as yf
except:
    !pip install yfinance --break-system-packages
    import yfinance as yf

import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Baixamos os dados do Ethereum
eth_data = yf.download('ETH-USD', start='2020-01-01', end='2023-12-31')
eth_data['Close'].plot(figsize=(14,10))
```

#Depois, podemos calcular o desvio padro destes dados.

```python
# Calculamos o desvio padro dos preos de fecho
eth_std = np.std(eth_data['Close'])
print('Desvio padro dos preos de fecho do Ethereum: ', eth_std)
```

##Alm disso, podemos ajustar os nossos dados a uma distribuio normal.

```python
# Ajustamos os dados a uma distribuio normal
mu, std = norm.fit(eth_data['Close'])
print('Mdia: ', mu)
print('Desvio padro: ', std)
```

##Por fim, podemos gerar uma distribuio normal a partir dos nossos dados e visualiz-la.

```python
import matplotlib.pyplot as plt

# Geramos uma distribuio normal
data_normal = np.random.normal(mu, std, 10000)

# Criamos um histograma para visualizar a distribuio
plt.hist(data_normal, bins=30, density=True, alpha=0.6, color='g')
plt.show()
```
